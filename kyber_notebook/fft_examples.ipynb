```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 20000          # Number of values to generate
frequency = 5                # Frequency of the sine wave (Hz)
sampling_rate = 20000        # Samples per second (Hz)
duration = num_samples / sampling_rate  # Duration in seconds

# Generate time values
t = np.linspace(0, duration, num_samples, endpoint=False)

# Generate sine wave values
sine_wave = np.sin(2 * np.pi * frequency * t)


plt.figure()
plt.plot(sine_wave)
plt.show()
plt.figure()
spectrum =np.fft.fft(sine_wave)
spectrum = abs(spectrum)
plt.plot(spectrum)
spectrum = abs(spectrum)
plt.show()
print(max(spectrum))
```

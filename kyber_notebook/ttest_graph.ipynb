```python
%matplotlib widget
import matplotlib.pylab as plt
from tqdm import trange
import numpy as np
from scipy.stats import ttest_ind
from scipy import stats
import time
import random
import os

# Mostrar una traza
def show_trace(trace):
    x = np.linspace(-10, 10, len(trace))  # Crear eje x con 2000 valores de -2 a 2
    plt.figure()
    plt.figure(figsize=(8, 5))
    plt.plot(x, trace, label="Nivel de confianza acumulado")
    plt.xlim(-10, 10)  # Ventana de valores del eje x
    plt.xlabel("Valor t-test")
    plt.ylabel("Nivel de confianza acumulado (CDF)")
    plt.axvline(x=4.5, color='green', linestyle='--', linewidth=1, label="TVLA umbral +t = 4.5")
    plt.axvline(x=-4.5, color='green', linestyle='--', linewidth=1, label="TVLA umbral -t = 4.5")
    plt.axhline(y=0.95, color='red', linestyle='--', linewidth=1, label="95% de confianza")
    plt.axhline(y=1, color='black', linestyle='--', linewidth=1, label="100% de confianza")
    plt.legend(loc="lower right")
    plt.show()
```

```python
valores = []
for i in range(1000):
    valores.append((-1000+i)/20)
for i in range(1000):
    valores.append(i/20)
p_lista = []

df = 1

for valor in valores:
    p_lista.append(1 - (2 * (1 - stats.t.cdf(abs(valor), df))))
```

```python
show_trace(p_lista)
```

```python
show_trace(p_lista)
```
